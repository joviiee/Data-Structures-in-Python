```python
import warnings

class Node:
    def __init__(self,val):
        self.val = val
        self.next = None

class LList:
    def __init__(self,val = None):
        if val:
            node = Node(val)
            self.head = node
        else:
            self.head = None

    def insert_at_beginning(self,val):
        node = Node(val)
        if self.head:
            node.next = self.head
            self.head = node
        else:
            self.head = node

    def print_list(self):
        current_node = self.head
        while current_node:
            print(current_node.val,end=',')
            current_node = current_node.next
        print('\n',end='')

    def insert_at_end(self,val):
        node = Node(val)
        if not self.head:
            self.head = node
        else:
            current_node = self.head
            while current_node.next:
                current_node = current_node.next
            current_node.next = node

    def insert_at_index(self,val,index):
        if index == 0:
            self.insert_at_beginning(val)
        else:
            ref = 0
            current_node = self.head
            while current_node.next and ref+1 != index:
                current_node = current_node.next
                ref+=1
            if not current_node.next and ref+1 == index:
                node = Node(val)
                current_node.next = node
                warnings.warn('Using insert_at_end() to insert node at the end can provide with performance advantage.')  
            elif not current_node.next:
                raise IndexError('Index out of range')
            else:
                node = Node(val)
                old_node = current_node.next
                current_node.next = node
                node.next = old_node

    def update_at_index(self,val,index):
        ref = 0
        current_node = self.head
        while current_node and ref != index:
            current_node = current_node.next
            ref+=1
        if current_node and ref == index:
            current_node.val = val
        else:
            raise IndexError('Index out of range')
        
    def remove_at_beginning(self):
        if self.head:
            self.head = self.head.next
    
    def remove_at_end(self):
        current_node =  self.head
        if current_node:
            while current_node.next.next:
                current_node = current_node.next
            current_node.next = None
        else:
            print('Llist is empty')

    def remove_at_index(self,index):
        if index == 0:
            self.remove_at_beginning()
        else:
            ref = 0
            current_node = self.head
            while current_node.next and ref+1 != index:
                current_node = current_node.next
                ref+=1
              
            if not current_node.next:
                raise IndexError('Index out of range')
            elif not current_node.next.next and ref+1 == index:
                    current_node.next = None
                    warnings.warn('Using remove_at_end() to remove node at the end can provide with performance advantage.')
            else:
                # old_node = current_node.next.next
                current_node.next = current_node.next.next

    def remove_value(self,val):
        if self.head.val == val:
            self.remove_at_beginning()
        else:
            current_node = self.head
            while current_node.next and current_node.next.val != val:
                current_node = current_node.next
              
            if not current_node.next:
                raise IndexError('Item not found')
            else:
                # old_node = current_node.next.next
                current_node.next = current_node.next.next

    def get_len(self):
        len = 0
        current_node = self.head
        while current_node:
            len+=1
            current_node = current_node.next
        print(len)
```
